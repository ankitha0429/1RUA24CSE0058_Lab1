```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer as si
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
df=pd.read_csv("Automobile - Automobile.csv")
df.head(20)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns="horsepower", inplace=True)
df.info()
```

```python
df.isnull().sum()
```

```python
df["cylinders"] = df["cylinders"].fillna(df["cylinders"].median())
df["displacement"] = df["displacement"].fillna(df["displacement"].median())
df["weight"] = df["weight"].fillna(df["weight"].median())
df["acceleration"] = df["acceleration"].fillna(df["acceleration"].median())
df.info()
df.isnull().sum()
```

```python
n=["cylinders","displacement","weight","acceleration"]
sc1=StandardScaler()
df[n]=sc1.fit_transform(df[n])
df.head(20)
```

```python
sc2=MinMaxScaler()
df[n]=sc2.fit_transform(df[n])
df.head()
```

Standardization makes more sense to the dataset as most of the values in the dataset are numeric and a lot of outliers can be seen

```python
df.to_csv("Auto_cleaned.csv",index=False)
```
